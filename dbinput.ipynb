```python
from sqlalchemy import create_engine, text, Table, Column, Integer, String, Date, MetaData
from sqlalchemy.dialects.mysql import VARCHAR, DATE
from sqlalchemy.orm import sessionmaker
import datetime
import os
from dotenv import load_dotenv
load_dotenv()

USERNAME = os.getenv("USERNAME")
PASSWORD = os.getenv("PASSWORD")
HOST = os.getenv("HOST")
DBNAME = os.getenv("DBNAME")
```

```python
# 2. Create engine and session
engine = create_engine(f"mysql+pymysql://{USERNAME}:{PASSWORD}@{HOST}/{DBNAME}", echo=True)
Session = sessionmaker(bind=engine)
session = Session()
metadata = MetaData()
```

```python
import pandas as pd

# 4. Read the CSV file
df = pd.read_csv("SYD.csv")
# Optional: Inspect the first few rows

df.head()
```

```python
# Convert 'DoB' and 'passport_expiry' columns to MySQL datetime format (YYYY-MM-DD)
df['DoB'] = pd.to_datetime(df['DoB'], errors='coerce').dt.strftime('%Y-%m-%d')
df['passport_expiry'] = pd.to_datetime(df['passport_expiry'], errors='coerce').dt.strftime('%Y-%m-%d') errors='coerce').dt.strftime('%Y-%m-%d')
df.head()
```

```python
try:
    # Attempt to connect and execute a simple query
    with engine.connect() as connection:
        result = connection.execute(text("SELECT 1"))
        print("Database connection successful:", result.scalar() == 1)
except Exception as e:
    print("Database connection failed:", e)
```

```python
# Insert data into tb_student
df.to_sql('tb_student', con=engine, if_exists='append', index=False)
print("Data inserted into tb_student table successfully.")

with engine.connect() as conn:
    # Fetch and print the first 5 rows from tb_student
    result = conn.execute(text("SELECT * FROM tb_student LIMIT 5"))
    for row in result:
        print(row)
```
